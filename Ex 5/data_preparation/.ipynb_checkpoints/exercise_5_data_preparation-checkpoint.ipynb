```python
import pandas as pd 
import numpy as np
import time
import datetime
```

```python
df = pd.read_csv('13_counties_stations_4_months_data.csv', index_col=0, parse_dates=True, infer_datetime_format=True)
df.head()
```

```python
df1 = df.groupby(['stations_id', 'altitude', 'latitude', 'longitude', 'name', 'state'])['r1'].sum()
```

```python
df1 = pd.DataFrame(df1)
```

```python
df1.reset_index(level=0, inplace=True)
```

```python
 data = df1[['stations_id','r1', 'altitude', 'latitude', 'longitude', 'name', 'state']]
```

```python
data
```

```python
data.rename(columns={'r1': 'val'}, inplace=True)
```
