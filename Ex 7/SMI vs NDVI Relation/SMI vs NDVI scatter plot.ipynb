```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.cbook as cbook
import seaborn as sns
import scipy
from adjustText import adjust_text
plt.style.use('seaborn')
%matplotlib inline
```

```python
df = pd.read_excel('SMI_NDVI_Value.xlsx', engine='openpyxl')
df.head()
```

```python
roi = df['ROI']
smi_16_may = df['SMI 16th May']
ndvi_10_may = df['NDVI 10th May']
```

```python
slope_may, intercept_may, r_value_may, p_value_may, std_err_may = scipy.stats.linregress(smi_16_may, ndvi_10_may)
print(slope_may, intercept_may, r_value_may, p_value_may, std_err_may)
r_value_square_may= r_value_may * r_value_may
print(r_value_square_may)
```

```python
fig = plt.figure(dpi = 100, figsize = (10, 5), clear = True)
plt.xlabel("SMI 16th May")
plt.ylabel("NDVI 10th May ")

plt.scatter(smi_16_may, ndvi_10_may, cmap = 'winter',  edgecolor = 'black', linewidth = 1, alpha = 0.75)

m, b = np.polyfit(smi_16_may, ndvi_10_may,  1)
plt.plot(smi_16_may, m*smi_16_may+b, color='red', label = "r_square = 0.36, standard_error = 9.16 p_value = 0.211")

#for i, txt in enumerate(station):
    #plt.annotate(txt, (val[i], altitude[i]), size=9)
    
texts = [plt.text(smi_16_may[i], ndvi_10_may[i], '%s' %txt) for  i, txt in enumerate(roi) ]
adjust_text(texts)

plt.title("NDVI vs SMI in May for ROI Olpe")
plt.legend(loc="upper right")

fig.savefig('ndvi_vs_smi_may.png')

plt.show()
```
